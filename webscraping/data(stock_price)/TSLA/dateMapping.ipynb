```python
import csv
from datetime import datetime, timedelta

close_prices_dict = {}

csv_filename = "stock_price.csv"

with open(csv_filename, "r") as file:
    reader = csv.reader(file)
    next(reader)  # Skip the header row
    for row in reader:
        date = row[0]  # Date is the first column
        close_price = float(row[4])  # Close price is the fifth column (index 4)
        close_prices_dict[date] = close_price

print(close_prices_dict)

print(list(close_prices_dict)[-1])
```

```python
# Assign the closing price of the marketClosedDate to the closing price of the closest future stock open date
# For example, the market was closed on 2010-07-03, however, we will assign its closing price to the closing price of 2010-07-06

from datetime import datetime, timedelta

def date_range(start_date, end_date):
    dates = []
    current_date = start_date
    while current_date <= end_date:
        dates.append(current_date.strftime('%Y-%m-%d'))
        current_date += timedelta(days=1)
    return dates

print

start_date = datetime.strptime('2010-05-10', '%Y-%m-%d')
end_date = datetime.strptime(list(close_prices_dict)[-1], '%Y-%m-%d')

print(start_date)
print(end_date)

all_dates = date_range(start_date, end_date)

marketClosedDates = [date for date in all_dates if date not in list(close_prices_dict.keys())]

print(marketClosedDates)
print(close_prices_dict)
```

```python
dateMapping = {}

def date_mapping(marketClosedDates):
    for closedDate in marketClosedDates:
        for date in list(close_prices_dict.keys()):
            dateMapping[date] = date
            if datetime.strptime(date, '%Y-%m-%d') > datetime.strptime(closedDate, '%Y-%m-%d'):
                dateMapping[closedDate] = date
                break

date_mapping(marketClosedDates)
```

```python
#sort the dictionary according to the dates

import datetime
sorted_items = sorted(dateMapping.items(), key=lambda x: datetime.datetime.strptime(x[0], '%Y-%m-%d'))

dateMappingSorted = dict(sorted_items)

print(dateMappingSorted)
```

```python
import csv

csv_filename = "dateMapping.csv"

with open(csv_filename, mode='w', newline='') as file:
    writer = csv.writer(file)
    
    # Write the header row
    writer.writerow(['Date', 'Close'])
    
    # Write each key-value pair to the CSV file
    for date, dateMappings in dateMappingSorted.items():
        writer.writerow([date, dateMappings])

print(f"CSV file '{csv_filename}' has been created successfully.")
```
