```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import numpy as np
import datetime
from ast import literal_eval
from google.colab import drive
import math

drive.mount('/content/drive')

file_path = '/content/drive/My Drive/Word_Embedding.csv'

df = pd.read_csv(file_path)

df.describe()
df.info()

df = df.fillna(0)
df['Dates'] = pd.to_datetime(df['Dates'])

# print(df["Dates"])

plt.figure(figsize = (16, 6))
plt.plot(df["Dates"], df["Percentage_Change"], marker = 'o', linestyle = '-', markersize = 1, color = 'tab:red')
plt.title("Tesla Closed Price Percentage Change Plot")
plt.xlabel("Date")
plt.ylabel("Percentage Change")
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.grid(True)

plt.show()

df['Dates'] = df['Dates'].apply(lambda x: x.toordinal())

X = df['Dates'].values.reshape(-1, 1)
print(X)
y = df['Percentage_Change'].values.reshape(-1, 1)
print(y)

import numpy as np
from sklearn.model_selection import TimeSeriesSplit
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error

def ts_cross_val(series, p_values, d_values, q_values, n_splits=5):
    size = int(len(series) / n_splits)
    mse_scores = []
    best_mse = 99999999
    best_para = None

    for p in p_values:
        for d in d_values:
            for q in q_values:
                predictions = []
                for t in range(1, n_splits):
                    history = series[:size * t]
                    model = ARIMA(history, order=(p,d,q))
                    model_fit = model.fit()
                    yhat = model_fit.forecast(steps=size)
                    predictions.extend(yhat)
                mse = mean_squared_error(series[size:size+len(predictions)], predictions)
                if mse < best_mse:
                  best_mse = mse
                  best_para = (p, d, q)
                mse_scores.append((p, d, q, mse))
                print('ARIMA(%s,%s,%s) MSE=%.3f' % (p,d,q,mse))
    return best_para, best_mse

# Splitting the dataset
split_point = int(len(df) * 0.8)
training_set = df['Percentage_Change'][:split_point].tolist()
testing_set = df['Percentage_Change'][split_point:].tolist()

# Setup for ARIMA parameters to explore
p_values = range(0, 6)
d_values = range(0, 3)
q_values = range(0, 6)

# Perform time series cross-validation to find the best parameters
best_parameters, best_mse = ts_cross_val(training_set, p_values, d_values, q_values)
```

```python
# Train the ARIMA model on the entire training set with the best parameters
print(best_parameters)
model = ARIMA(training_set, order = best_parameters)
model_fit = model.fit()

# Forecast the test set
forecasts = model_fit.forecast(steps=len(testing_set))

# Evaluate performance
mse = mean_squared_error(testing_set, forecasts)
print(f"Test MSE: {mse}")
```
