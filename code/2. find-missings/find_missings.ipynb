```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from bisect import bisect_left

stock_price = pd.read_csv('stock_price.csv')
merged_df = pd.read_csv('merged_tesla_headlines.csv')
stock_dates = stock_price['Date'].tolist()
print(merged_df.head(10))
```

```python
def find_nearest_future_date(target, dates):
    """Find the nearest future date in 'dates' to the 'target' date."""
    pos = bisect_left(dates, target)
    if pos == len(dates):
        # Target is later than any date we have, return None or keep original
        return None
    return dates[pos]
merged_df = merged_df[merged_df['Dates'] >= '2010-06-29']
merged_df['MappedDates'] = merged_df['Dates'].apply(lambda d: find_nearest_future_date(d, stock_dates) if d not in stock_dates else d)
```

```python
print(merged_df[merged_df['Headlines'] == 'Tesla Breeding Devotees Even Before Electric Cars Are Built'])
```

```python
all_dates = set(stock_price['Date'])
unique_dates = set(merged_df['MappedDates'])
```

```python
missing_dates = []

for date in all_dates:
    if date not in unique_dates:
        missing_dates.append(date)

missing_dates.sort()

print(missing_dates)
```
