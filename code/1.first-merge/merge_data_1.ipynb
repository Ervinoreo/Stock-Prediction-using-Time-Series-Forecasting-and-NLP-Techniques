```python
tesla_2010 = pd.read_csv('tesla_google_2010.csv')
tesla_2011 = pd.read_csv('tesla_google_2011.csv')
tesla_2012 = pd.read_csv('tesla_google_2012.csv')
tesla_2013 = pd.read_csv('tesla_google_2013.csv')
tesla_2014 = pd.read_csv('tesla_google_2014.csv')
tesla_2015 = pd.read_csv('tesla_google_2015.csv')
tesla_2016 = pd.read_csv('tesla_google_2016.csv')
tesla_2017 = pd.read_csv('tesla_google_2017.csv')
tesla_2018 = pd.read_csv('tesla_google_2018.csv')
tesla_2019 = pd.read_csv('tesla_google_2019.csv')
tesla_2020 = pd.read_csv('tesla_google_2020.csv')
tesla_2021 = pd.read_csv('tesla_google_2021.csv')
tesla_2022 = pd.read_csv('tesla_google_2022.csv')
tesla_2023 = pd.read_csv('tesla_google_2023.csv')
tesla_blog = pd.read_csv('tesla_blog_news.csv')
```

```python
df_list = [tesla_2010, tesla_2011, tesla_2012, tesla_2013, tesla_2014, tesla_2015, tesla_2016, tesla_2017, tesla_2018, tesla_2019, tesla_2020, tesla_2021, tesla_2022, tesla_2023]
```

```python
def convert_to_datetime(dates):
    list = dates.split(", ")
    month_day = list[-2]
    year = list[-1]
    date = f"{month_day} {year}"
    return datetime.strptime(date, "%B %d %Y")

tesla_blog['Dates'] = tesla_blog['Dates'].apply(convert_to_datetime)
```

```python
print(tesla_blog.head(10))
```

```python
combined_df = pd.concat(df_list, ignore_index=True)[['Headlines', 'Dates', 'Descriptions']]
```

```python
def convert_to_datetime_2(dates):
    list = dates.split(" ")
    day = list[0]
    month = list[1][:3]
    year = list[2]
    date = f"{day} {month} {year}"
    return datetime.strptime(date, "%d %b %Y")

combined_df['Dates'] = combined_df['Dates'].apply(convert_to_datetime_2)
```

```python
print(combined_df.head(3))
```

```python
# Concat google news and tesla blog
final_df = pd.concat([combined_df, tesla_blog], ignore_index = True)

# Sort data by dates
final_df.sort_values(by=['Dates'], inplace=True)
```

```python
# final_df to csv
final_df.to_csv('merged_tesla_headlines.csv', index = False)
```
