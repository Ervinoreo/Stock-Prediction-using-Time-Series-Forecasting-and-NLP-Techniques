```python
# Text preprocessing
!pip install gensim nltk
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('punkt')
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))
# Remove Punctuation
df['Combined_Desc'] = df['Combined_Desc'].apply(lambda x: re.sub(r'[^\w\s]', '', x))

# Remove stopwords & Lowercasing
def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(text)
    filtered_text = [word.lower() for word in word_tokens if word.lower() not in stop_words]
    return ' '.join(filtered_text)

df['Combined_Desc'] = df['Combined_Desc'].apply(remove_stopwords)
```

```python
# Lemmatization
lmtzr = WordNetLemmatizer()

def lemmatize_text(text):
    tokens = word_tokenize(text)
    lemmatized_tokens = [lmtzr.lemmatize(word) for word in tokens]
    return ' '.join(lemmatized_tokens)

df['lemmatized_text'] = df['Combined_Desc'].apply(lemmatize_text)
df['lemmatized_text'][0]
```

```python
# TF-IDF
tfidf = TfidfVectorizer(max_features=100)
tfidf_matrix = tfidf.fit_transform(df['lemmatized_text'])
df['embeddings'] = list(tfidf_matrix.toarray())
print(df.head(3))
# result_df.to_csv('Word_Embedding.csv', index=False)
```
