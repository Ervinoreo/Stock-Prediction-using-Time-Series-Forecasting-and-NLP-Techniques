```python
data = pd.read_csv("tesla_data_no_perc_no_combined.csv")
df = pd.DataFrame(data)
```

```python
print(df.head(3))
```

```python
merged_df = df.groupby('Dates').agg({
    'Headlines': ' '.join,
    'Descriptions': ' '.join,
    'Closed Price': 'mean'
}).reset_index()
```

```python
print(merged_df.head(3))
```

```python
merged_df["Combined_Desc"] = merged_df['Headlines'] + ' ' + merged_df['Descriptions']
final_df = merged_df.drop(['Headlines', 'Descriptions'], axis=1)
```

```python
print(final_df)
```

```python
final_df["Percentage_Change"] = final_df['Closed Price'].pct_change() * 100
print(final_df.head(3))
```

```python
final_df.to_csv('tesla_final.csv', index=False)
```
